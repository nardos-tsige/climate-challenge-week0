## Interpretation of Summary Statistics - Ethiopia Climate Data (2015-2026)

### Temperature Analysis (T2M - Mean Daily Temperature)

- **Average Temperature:** 16.07°C over the entire period
- **Temperature Range:** 10.03°C (minimum) to 21.53°C (maximum)
- **Standard Deviation:** 1.90°C, indicating moderate seasonal variation
- **Median vs Mean:** 16.04°C vs 16.07°C - nearly identical, suggesting symmetrical temperature distribution

### Maximum Temperature (T2M_MAX)

- **Average Daily Max:** 23.20°C
- **Extreme Heat Days:** Maximum recorded at 30.93°C
- **Range:** 15.65°C to 30.93°C (15.28°C spread)

### Minimum Temperature (T2M_MIN)

- **Average Daily Min:** 10.23°C
- **Cool Nights:** Minimum recorded at 1.17°C
- **Range:** 1.17°C to 15.68°C

### Daily Temperature Range (T2M_RANGE)

- **Average Daily Range:** 12.97°C
- **Range Spread:** 3.56°C to 23.24°C
- **Implication:** Large diurnal temperature variation typical of highland climates (Ethiopian highlands)

### Precipitation Analysis (PRECTOTCORR)

- **Average Daily Rainfall:** 3.63 mm/day
- **Rainfall Pattern:** Highly skewed - 75% of days have ≤4.58 mm, but maximum reaches 82.30 mm
- **Zero Rainfall Days:** Minimum of 0.00 mm indicates many dry days
- **Standard Deviation:** 6.29 mm (higher than mean, confirming skewness toward dry conditions)

### Humidity (RH2M)

- **Average Relative Humidity:** 68.41%
- **Range:** 14.42% to 91.93%
- **Interpretation:** Generally humid conditions, typical of tropical climate

### Wind Speed (WS2M)

- **Average Wind Speed:** 1.98 m/s (light breeze)
- **Maximum Gust:** 7.49 m/s (moderate breeze)
- **Generally calm conditions** with occasional stronger winds

### Atmospheric Pressure (PS)

- **Average Pressure:** 77.04 kPa
- **Very stable** (standard deviation only 0.10 kPa)
- **Elevation indicator:** Lower than sea-level pressure (101.3 kPa), confirming Ethiopia's highland topography

### Key Observations for COP32 Negotiations:

1. **Temperature Trend:** Ethiopia shows moderate temperatures (avg 16°C) but with extremes reaching 31°C, indicating warming pressures

2. **Precipitation Vulnerability:** Rainfall is highly variable (0-82 mm/day), with many dry days - suggests drought risk and water insecurity

3. **Diurnal Range:** Large day-night temperature swings (avg 13°C) affects agriculture and energy demand

4. **Data Quality:** No obvious missing value issues with 4,108 complete observations from 2015-2026

```python
from scipy import stats

# Variables to check for outliers
variables = ['T2M', 'T2M_MAX', 'T2M_MIN', 'PRECTOTCORR', 'RH2M', 'WS2M', 'WS2M_MAX']

# Store outlier counts
outlier_counts = {}

for var in variables:
    # Drop NaN values for Z-score calculation
    clean_data = df[var].dropna()
    z_scores = np.abs(stats.zscore(clean_data))
    outlier_counts[var] = (z_scores > 3).sum()
    print(f"{var}: {outlier_counts[var]} outliers | {outlier_counts[var]/len(clean_data)*100:.2f}% of data")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv(r"C:\Users\HP\climate-challenge-week0\ethiopia.csv")

df['Country'] = "Ethiopia"
df['Date'] = pd.to_datetime(df['YEAR'] * 1000 + df['DOY'], format="%Y%j")
df['Month'] = df['Date'].dt.month

# Replace -999 with NaN
df.replace(-999, np.nan, inplace=True)

# Outlier detection (no scipy needed)
variables = ['T2M', 'T2M_MAX', 'T2M_MIN', 'PRECTOTCORR', 'RH2M', 'WS2M', 'WS2M_MAX']
for var in variables:
    clean_data = df[var].dropna()
    z_scores = np.abs((clean_data - clean_data.mean()) / clean_data.std())
    outliers = (z_scores > 3).sum()
    print(f"{var}: {outliers} outliers ({outliers/len(clean_data)*100:.2f}%)")

# Fill missing values
weather_cols = ['T2M', 'T2M_MAX', 'T2M_MIN', 'T2M_RANGE', 'PRECTOTCORR', 'RH2M', 'WS2M', 'WS2M_MAX', 'PS', 'QV2M']
df[weather_cols] = df[weather_cols].ffill()

# Save cleaned data
os.makedirs("../data", exist_ok=True)
df.to_csv("../data/ethiopia_clean.csv", index=False)

print("\n DONE! Cleaned data saved to ../data/ethiopia_clean.csv")
print(f" Shape: {df.shape}")
print(f" Date range: {df['Date'].min()} to {df['Date'].max()}")
```
