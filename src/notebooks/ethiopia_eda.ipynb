```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv(r"C:\Users\HP\climate-challenge-week0\ethiopia.csv")

df['Country'] = "Ethiopia"
df['Date'] = pd.to_datetime(df['YEAR'] * 1000 + df['DOY'], format="%Y%j")
df['Month'] = df['Date'].dt.month

# Replace -999 with NaN
df.replace(-999, np.nan, inplace=True)

# Outlier detection (no scipy needed)
variables = ['T2M', 'T2M_MAX', 'T2M_MIN', 'PRECTOTCORR', 'RH2M', 'WS2M', 'WS2M_MAX']
for var in variables:
    clean_data = df[var].dropna()
    z_scores = np.abs((clean_data - clean_data.mean()) / clean_data.std())
    outliers = (z_scores > 3).sum()
    print(f"{var}: {outliers} outliers ({outliers/len(clean_data)*100:.2f}%)")

# Fill missing values
weather_cols = ['T2M', 'T2M_MAX', 'T2M_MIN', 'T2M_RANGE', 'PRECTOTCORR', 'RH2M', 'WS2M', 'WS2M_MAX', 'PS', 'QV2M']
df[weather_cols] = df[weather_cols].ffill()

# Save cleaned data
os.makedirs("../data", exist_ok=True)
df.to_csv("../data/ethiopia_clean.csv", index=False)

print("\n DONE! Cleaned data saved to ../data/ethiopia_clean.csv")
print(f" Shape: {df.shape}")
print(f" Date range: {df['Date'].min()} to {df['Date'].max()}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data you already saved
df = pd.read_csv("../data/ethiopia_clean.csv")

# Convert Date back to datetime if needed
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

print(f"Loaded {len(df)} rows")
print(f"Date range: {df['Date'].min()} to {df['Date'].max()}")
df.head()
```

```python
# Monthly average temperature
monthly_temp = df.groupby(df['Date'].dt.to_period('M'))['T2M'].mean()

plt.figure(figsize=(14, 5))
plt.plot(monthly_temp.index.astype(str), monthly_temp.values, color='red', linewidth=1.5)

# Annotate warmest month
warmest = monthly_temp.idxmax()
warmest_val = monthly_temp.max()
plt.annotate(f'Warmest: {warmest_val:.1f}°C', 
             xy=(str(warmest), warmest_val),
             xytext=(10, 10), textcoords='offset points',
             arrowprops=dict(arrowstyle='->', color='red'))

# Annotate coolest month
coolest = monthly_temp.idxmin()
coolest_val = monthly_temp.min()
plt.annotate(f'Coolest: {coolest_val:.1f}°C',
             xy=(str(coolest), coolest_val),
             xytext=(10, -20), textcoords='offset points',
             arrowprops=dict(arrowstyle='->', color='blue'))

plt.title('Monthly Average Temperature in Ethiopia (2015-2026)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n📊 Summary:")
print(f"   Warmest month: {warmest} ({warmest_val:.1f}°C)")
print(f"   Coolest month: {coolest} ({coolest_val:.1f}°C)")
```

```python
# Monthly total precipitation
monthly_rain = df.groupby(df['Date'].dt.to_period('M'))['PRECTOTCORR'].sum()

plt.figure(figsize=(14, 5))
bars = plt.bar(monthly_rain.index.astype(str), monthly_rain.values, color='blue', alpha=0.7)

# Annotate peak rainy months
top3 = monthly_rain.nlargest(3)
for month in top3.index:
    idx = list(monthly_rain.index).index(month)
    plt.annotate(f'{top3[month]:.0f}mm', 
                 xy=(idx, top3[month]),
                 xytext=(0, 10), textcoords='offset points',
                 ha='center', fontsize=9)

plt.title('Monthly Total Precipitation in Ethiopia (2015-2026)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Precipitation (mm)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print(f"\nPeak rainy months:")
for month, rain in top3.items():
    print(f"   {month}: {rain:.1f} mm")
```

```python
# Select numeric columns
corr_cols = ['T2M', 'T2M_MAX', 'T2M_MIN', 'PRECTOTCORR', 'RH2M', 'WS2M', 'QV2M', 'PS']
corr_matrix = df[corr_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', square=True)
plt.title('Correlation Matrix of Climate Variables - Ethiopia', fontsize=14)
plt.tight_layout()
plt.show()

# Find strongest correlations
corr_pairs = corr_matrix.unstack().sort_values(ascending=False)
corr_pairs = corr_pairs[corr_pairs < 0.999]  # Remove self-correlations

print("\nTop 3 Strongest Correlations:")
for i in range(3):
    var1, var2 = corr_pairs.index[i]
    print(f"   {i+1}. {var1} vs {var2}: {corr_pairs.values[i]:.3f}")
```
